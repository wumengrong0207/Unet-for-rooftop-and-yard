```python
#在arcgis中导出的标签文件和图像文件，标签文件自动去除了没有目标物的图，用这段代码，将与标签文件对应的有目标物的image文件筛选出来
import os
import shutil

def copy_files_to_new_folders(image_folder, label_folder, output_image_folder, output_label_folder):
    """
    将 image 和 label 文件夹中的文件复制到新的文件夹中。
    
    参数：
        image_folder (str): 存储图像文件的文件夹路径。
        label_folder (str): 存储标签文件的文件夹路径。
        output_image_folder (str): 目标图像文件夹路径。
        output_label_folder (str): 目标标签文件夹路径。
    """
    # 确保目标文件夹存在
    os.makedirs(output_image_folder, exist_ok=True)
    os.makedirs(output_label_folder, exist_ok=True)

    # 获取 image 和 label 文件夹中的所有文件
    image_files = sorted([f for f in os.listdir(image_folder) if os.path.splitext(f)[1] == '.TIF'])
    label_files = sorted([f for f in os.listdir(label_folder) if os.path.splitext(f)[1] == '.TIF'])

    for label_file in label_files:
        # 假设图片文件的命名规则是：label文件去掉 'labels' 替换为 'images'
        image_file = label_file.replace('label', 'image')
        # 如果在 image 文件夹中找到对应的标签文件，则进行复制
        if image_file in image_files:
            img_path = os.path.join(image_folder, image_file)
            label_path = os.path.join(label_folder, label_file)

            # 复制到新的文件夹
            shutil.copy(img_path, os.path.join(output_image_folder, image_file))
            shutil.copy(label_path, os.path.join(output_label_folder, label_file))

           

# 输入和输出文件夹路径
image_folder = r"E:\rural rooftop and courtyard\dataset_from_GIS\image_shandong_version2"
label_folder = r"E:\rural rooftop and courtyard\dataset_from_GIS\label_shandong_version2"
output_image_folder = r"E:\rural rooftop and courtyard\dataset\image_shandong_roofandyard_version2"
output_label_folder = r"E:\rural rooftop and courtyard\dataset\label_shandong_roofandyard-version2"

# 复制匹配的图像和标签文件到新文件夹
copy_files_to_new_folders(image_folder, label_folder, output_image_folder, output_label_folder)
print("复制图像文件完成")
```
